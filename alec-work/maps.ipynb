```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from vacay_key import g_key
```

```python
# reading in lat lon of city capitols
capitols_df = pd.read_csv("../resources/state_cap_lat_lon.csv",index_col=0)
capitols_df
```

```python
# reading in cities csv from part 1
most_pop_2018_df = pd.read_csv("output/2018_top_alc_by_state.csv",index_col=0)
most_pop_2018_df
```

```python
print(len(capitols_df["STATE CODE"]))
print(len(most_pop_2018_df["Type of Beverage"]))
```

```python
cap_most_pop = pd.merge(capitols_df, most_pop_2018_df,left_index=True, right_index=True)
cap_most_pop
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=g_key)
```

```python
# testing out my fig
fig = gmaps.figure()
```

```python
beer_df = cap_most_pop.loc[cap_most_pop["Type of Beverage"] == "Beer"]
wine_df = cap_most_pop.loc[cap_most_pop["Type of Beverage"] == "Wine"]
spirits_df = cap_most_pop.loc[cap_most_pop["Type of Beverage"] == "Spirits"]
```

```python
locations_beer = beer_df[["LAT", "LON"]].astype(float)
locations_wine = wine_df[["LAT", "LON"]].astype(float)
locations_spirits = spirits_df[["LAT", "LON"]].astype(float)
locations_all = cap_most_pop[["LAT", "LON"]].astype(float)
```

```python

symbol_layer_beer = gmaps.symbol_layer(
        locations_beer,fill_color="green",stroke_color="green",scale=4)
symbol_layer_wine = gmaps.symbol_layer(
        locations_wine,fill_color="red",stroke_color="red",scale=4)
symbol_layer_spirits = gmaps.symbol_layer(
        locations_spirits,fill_color="blue",stroke_color="blue",scale=4)

fig.add_layer(symbol_layer_beer)
fig.add_layer(symbol_layer_wine)
fig.add_layer(symbol_layer_spirits)
fig
```
