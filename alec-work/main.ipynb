```python
import pandas as pd
import numpy as np
```

```python
# reading in niaaa csv data
niaaa_path = "../resources/niaaa_data_normalized.csv"
niaaa_df = pd.read_csv(niaaa_path)
niaaa_df
```

```python
# renaming some columns for ease
niaaa_df_col = niaaa_df.rename(columns={"Year (4-digit calendar year)" : "Year",
                                       "Geographic ID code (FIPS code, see specification below)" : "Geo ID Code",
                                       "Type of beverage (see specification below)" : "Beverage Type Code"},
                                       )
niaaa_df_col
```

```python
niaaa_df_col.dtypes
```

```python
# normalizing object types to float
gal_of_bev_num = pd.to_numeric(niaaa_df_col["Gallons of beverage"], errors='coerce')
```

```python
# adding column back into the df
niaaa_df_col["Gallons of beverage"] = gal_of_bev_num

# fill na with zero
niaaa_df_col_drop = niaaa_df_col.fillna(0)
niaaa_df_col_drop
```

```python
# normalizing columns by the millions
niaaa_df_col_drop["Gallons of beverage"] = niaaa_df_col_drop["Gallons of beverage"] / 1000000
niaaa_df_col_drop["Gallons of ethanol (absolute alcohol)"] = niaaa_df_col_drop["Gallons of ethanol (absolute alcohol)"] / 1000000
niaaa_df_col_drop["Population (age 14 and older)"] = niaaa_df_col_drop["Population (age 14 and older)"] / 1000000
niaaa_df_col_drop["Population (age 21 and older)"] = niaaa_df_col_drop["Population (age 21 and older)"] / 1000000
niaaa_df_col_drop
```

```python
# creating df for us as a whole
us_df = niaaa_df_col_drop.loc[niaaa_df_col_drop["State_Region"] == "United States"]
us_df
```

```python
# grouping by year looking at US
year_group_us = us_df.groupby("Year")
year_group_us
```

```python
# creating summary stats by year

# Gallons of beverage *gob*, Gallons of ethanol (absolute alcohol) *goe*, Population (age 14 and older) *pop_14*,
# Gallons of ethanol per capita age 14 and older *goe_pcap_14*, 
# Population (age 21 and older) *pop_21*, Gallons of ethanol per capita age 21 and older *goe_pcap_21*

year_gob_sum = year_group_us["Gallons of beverage"].sum()
year_goe_sum = year_group_us["Gallons of ethanol (absolute alcohol)"].sum()
year_pop_14_sum = year_group_us["Population (age 14 and older)"].sum() / 4
year_goe_pcap_14_sum = year_group_us["Gallons of ethanol per capita age 14 and older"].sum()
year_pop_21_sum = year_group_us["Population (age 21 and older)"].sum() / 4
year_goe_pcap_21_sum = year_group_us["Gallons of ethanol per capita age 21 and older"].sum()
```

```python
# group by year df
us_gob_year_df = pd.DataFrame(year_gob_sum)
us_geo_year_df = pd.DataFrame(year_goe_sum)
us_pop_14_year_df = pd.DataFrame(year_pop_14_sum)
us_goe_pcap_14_year_df = pd.DataFrame(year_goe_pcap_14_sum)
us_pop_21_year_df = pd.DataFrame(year_pop_21_sum)
us_goe_pcap_21_year_df = pd.DataFrame(year_goe_pcap_21_sum)

print(len(us_gob_year_df))
print(len(us_geo_year_df))
print(len(us_pop_14_year_df))
print(len(us_goe_pcap_14_year_df))
print(len(us_goe_pcap_21_year_df))

```

```python
# merge
merge_us_1 = pd.merge(us_gob_year_df,us_geo_year_df,left_index=True, right_index=True)
merge_us_2 = pd.merge(merge_us_1, us_pop_14_year_df,left_index=True, right_index=True)
merge_us_3 = pd.merge(merge_us_2, us_goe_pcap_14_year_df,left_index=True, right_index=True)
merge_us_4 = pd.merge(merge_us_3, us_pop_21_year_df,left_index=True, right_index=True)
us_year_df = pd.merge(merge_us_4, us_goe_pcap_21_year_df,left_index=True, right_index=True)
us_year_df
```

```python
# output data on stats by year
us_year_path = "output/niaaa_us_yearly_consumption_stats_clean.csv"
us_year_df.to_csv(us_year_path,index=True)
```
